# Ejercicio 1
Se le ha encargado la validación del dígito verificador del RUT que ingresan las personas que solicitan permisos individuales durante la cuarentena. Para ello, cada persona debe ingresar su RUT separado del dígito verificador. Se deberá validar si el dígito verificador corresponde al ingresado, desplegar un mensaje en pantalla que indique “RUT correcto” y con ello proseguir el proceso de solicitud de permiso.

El dígito verificador se valida de acuerdo a lo siguiente:
* Multiplicar cada dígito del RUT (sin incluir dígito verificador) inverso por la secuencia 2, 3,
4, 5, 6, 7 y luego vuelve a repetirse: ejemplo para RUT: 17355042:
* RUT invertido: 2 4 0 5 5 3 7 1
* 2x2, 3x4, 4x0, 5x5, 6x5, 7x3, 2x7, 3x1
* Sumar lo anterior: 2x2+3x4+4x0+5x5+6x5+7x3+2x7+3x1=4+12+4+25+30
+21+14+3= 109
* Obtener la división entera de 109 / 11: 9
* Multiplicar 9 x 11: 99
* Efectuar la resta: 109 - 99 = 10
* Efectuar la resta: 11 - 10 = 1
* El dígito verificador es 1

Cálculo de dígito verificador de RUT en Chile

Escriba un programa en Python que valide si el RUT es correcto o no.

Un ejemplo de ejecución es el siguiente (las respuestas del usuario están en negrillas):

Se validará a continuación su RUT
Ingrese el RUT sin el dígito verificador: **19488371**
Ingrese el dígito verificador de su RUT: **2**
El RUT 19488371-2 no es correcto

Se validará a continuación su RUT
Ingrese el RUT sin el dígito verificador: **19488371**
Ingrese el dígito verificador de su RUT: **4**
El RUT 19488371-4 es correcto
